Problem 1: Unique Words in a Sentence
Problem Statement
Given a sentence, identify all unique words and display:
The count of unique words
The unique words themselves


```python
i = input("enter the sentence: ")
y = i.split()
unique = []
for j in y:
    if y.count(j) == 1:
        unique.append(j)
print("Count:", len(unique))
print("Unique words:", unique)
```

Problem 2: Highest Salary from Employee Data
Problem Statement
Store employee names and their salaries, and determine which employee has the highest salary.
Hint
Use a dictionary where the key is the employee name and the value is the salary.
Compare salary values to find the highest.


```python
d={}
n=int(input("enter the total number of employees"))
for i in range(n):
    name=input("enter the name of the employee:")
    salary=int(input("enter the salary of the employee:"))
    d[name]=salary

    max_salary=max(d.values())
    print("maximum salary ",max_salary)
```

Problem 4: Count Products Above a Price Threshold
Problem Statement
Count how many products have a price greater than 1000.


```python
prices = [450, 1200, 899, 1500, 300]
count=0
for i in range(len(prices)):
  if prices[i]>1000 :
    count=count+1
print("number of prices above 1000 ",count)
```

Problem 5: Calculate Attendance Percentage
Problem Statement
Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).


```python
attendance = ["P", "P", "A", "P", "P","A","A"]
count=0
for i in range(len(attendance)):
  if attendance[i]=="P":
    count=count+1
    percentage=(count/len(attendance))*100
print("attendance percentage ",percentage)
```

Problem 6: Remove Duplicate Phone Numbers
Problem Statement
You are given a list of phone numbers that may contain duplicates. Write a Python program to remove duplicate phone numbers and display only the unique phone numbers.


```python
phone_numbers = ["9876543210", "9123456789", "9876543210", "9001122334", "9123456789"]
new = []
for num in phone_numbers:
    if num not in new:
        new.append(num)
print("Unique Phone Numbers are :", new)
```

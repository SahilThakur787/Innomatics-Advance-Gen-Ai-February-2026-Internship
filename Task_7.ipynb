Problem 1: Smart Light Controller
Problem Statement
Create a class to control a smart light that can be turned ON or OFF and display its status.


```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"
    def turn_on(self):
        self.status = "ON"
    def turn_off(self):
        self.status = "OFF"
    def display_status(self):
        print(self.name + " is " + self.status)
light_name = "Bedroom Light"
action = "ON"
light = SmartLight(light_name)
if action == "ON":
    light.turn_on()
elif action == "OFF":
    light.turn_off()
light.display_status()
```

Problem 3: Mobile Contact Record
Problem Statement
Create a class to store contact name and phone number and display contact information


```python
class Phone:
  def enter(self,name,number):
    self.name=name
    self.number=number
  def display(self):
    print("contact saved")
    print("name:",self.name)
    print("phone no:",self.number)
name=input("enter the name:")
phone_no=input("enter the phone number:")
y=Phone()
y.enter(name,phone_no)
y.display()
```

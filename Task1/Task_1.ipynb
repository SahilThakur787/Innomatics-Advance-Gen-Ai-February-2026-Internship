Problem Statement:
Analyze student results.
marks = [45, 78, 90, 33, 60]

Requirements:
A student passes if marks â‰¥ 50
Count the total number of pass students
Count the total number of fail students
Print the final result clearly


```python
marks=[45,78,90,33,60]
a=0
b=0
for i in marks:
  if i>=50:
    print("marks" ,i," pass")
    a+=1
  else:
    print("marks" ,i," fail")
    b+=1
print("total number of the passed student", a)
print("total number of the failed student", b)
```

Problem Statement:
Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

Requirements:
Count the number of "ERROR" entries
Print the total error count


```python
logs=["INFO", "ERROR", "WARNING", "ERROR"]
a='ERROR';
count=0
for i in logs:
  if i==a:
    count+=1;
print("total number of ERRORS IN THE LOGS ARE:",count)
```

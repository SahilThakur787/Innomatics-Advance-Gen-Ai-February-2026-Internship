```python
from langchain_core.prompts import PromptTemplate
from langchain_core.prompts import ChatPromptTemplate

print("Libraries imported successfully")
```

```python
def explain_topic_old(topic):
    return "Explain " + topic + " in simple terms for beginners"

topic = "Artificial Intelligence"

print("Input:", topic)
print("Output:", explain_topic_old(topic))
```

```python
template = "Explain {topic} in simple terms for beginners"

prompt_template = PromptTemplate(
    input_variables=["topic"],
    template=template
)

topic = "Artificial Intelligence"

print("Input:", topic)

result = prompt_template.format(topic=topic)

print("Generated Prompt:", result)
```

```python
multi_prompt = PromptTemplate(
    input_variables=["topic","audience","tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

topic="AI"
audience="beginners"
tone="friendly"

print("Input Values")
print("Topic:",topic)
print("Audience:",audience)
print("Tone:",tone)

result = multi_prompt.format(topic=topic,audience=audience,tone=tone)

print("\nGenerated Prompt:")
print(result)
```

```python
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic="Machine Learning"

print("Input Topic:",topic)

print("\nTeaching Style Output:")
print(teaching_template.format(topic=topic))

print("\nInterview Style Output:")
print(interview_template.format(topic=topic))

print("\nStory Style Output:")
print(story_template.format(topic=topic))
```

```python
chat_template = ChatPromptTemplate.from_messages([
("system","You are acting as a {role}"),
("user","Explain {topic}")
])

role="teacher"
topic="Neural Networks"

print("Input Role:",role)
print("Input Topic:",topic)

messages = chat_template.format_messages(role=role,topic=topic)

print("\nGenerated Chat Messages:")
print(messages)
```

```python
def validate_inputs(audience,tone):

    valid_audience=["beginner","intermediate","expert"]
    valid_tone=["formal","casual","fun"]

    if audience not in valid_audience:
        print("Invalid audience detected → setting default = beginner")
        audience="beginner"

    if tone not in valid_tone:
        print("Invalid tone detected → setting default = formal")
        tone="formal"

    return audience,tone


audience="kids"
tone="fun"

print("Original Input:",audience,tone)

audience,tone=validate_inputs(audience,tone)

print("Validated Output:",audience,tone)
```

```python
generator_template = PromptTemplate(
    input_variables=["topic","audience","tone","style"],
    template="Explain {topic} for {audience} in a {tone} tone using {style} style"
)

def generate_prompt(topic,audience,tone,style):

    audience,tone=validate_inputs(audience,tone)

    prompt=generator_template.format(
        topic=topic,
        audience=audience,
        tone=tone,
        style=style
    )

    return prompt


print("Input Values")
print("Topic: Neural Networks")
print("Audience: beginner")
print("Tone: fun")
print("Style: storytelling")

result=generate_prompt("Neural Networks","beginner","fun","storytelling")

print("\nGenerated Prompt:")
print(result)
```

```python
reuse_template = PromptTemplate(
    input_variables=["topic","audience"],
    template="Explain {topic} for {audience}"
)

tests=[
("AI","students"),
("Python","beginners"),
("Deep Learning","engineers"),
("Data Science","analysts"),
("NLP","researchers")
]

print("Reusability Test\n")

for topic,audience in tests:
    print("Input:",topic,",",audience)
    print("Output:",reuse_template.format(topic=topic,audience=audience))
    print()
```

```python
from langchain_core.prompts import PromptTemplate

# Complete pipeline demonstration
def run_pipeline(topic, audience, tone, style):

    print("="*60)
    print("PROMPT GENERATION PIPELINE")
    print("="*60)

    # Step 1 — User Input
    print("\nSTEP 1: User Input")
    print("Topic:", topic)
    print("Audience:", audience)
    print("Tone:", tone)
    print("Style:", style)

    # Step 2 — Input Validation
    print("\nSTEP 2: Input Validation")
    audience, tone = validate_inputs(audience, tone)
    print("Validated Audience:", audience)
    print("Validated Tone:", tone)

    # Step 3 — Prompt Template
    print("\nSTEP 3: Prompt Template Selection")

    template = PromptTemplate(
        input_variables=["topic","audience","tone","style"],
        template="Explain {topic} for {audience} in a {tone} tone using {style} style"
    )

    print("Template Structure:")
    print("Explain {topic} for {audience} in a {tone} tone using {style} style")

    # Step 4 — Prompt Generation
    print("\nSTEP 4: Dynamic Prompt Generation")

    generated_prompt = template.format(
        topic=topic,
        audience=audience,
        tone=tone,
        style=style
    )

    # Step 5 — Final Output
    print("\nSTEP 5: Final Output")
    print(generated_prompt)

    print("\nPipeline Completed Successfully")

    return generated_prompt


# Run pipeline example
run_pipeline(
    topic="Neural Networks",
    audience="beginner",
    tone="fun",
    style="storytelling"
)
```
